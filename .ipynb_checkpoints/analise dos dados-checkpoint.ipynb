Observamos que durante 2014 e 2018, não há aumentos significativos, sendo necessário analisar o porquê. A partir de 2018, há o início de uma crescente exponencial que se destaca a partir de 2021, o que também será cabível de análise

Hipóteses: 

- Tendência nacional (comparar com outros estados)
- Crescimento do Estado da paraíba
- Mudança de gestão do governo da paraíba


#### Hipótese 2: Crescimento Populacional e Valor Orçado

```python
df_soma_orcamento = df.groupby("exercicio")["valor_orcado"].sum().reset_index()

df_soma_orcamento.columns = ["Ano", "Somatorio_Valor_Orcado"]

df_combined = df_soma_orcamento.merge(popu_pb, on="Ano")

correlacao = df_combined[['Somatorio_Valor_Orcado', 'População']].corr()

print("Matriz de Correlação:")
correlacao
```

```python
correlacao, _ = pearsonr(df_combined['População'], df_combined['Somatorio_Valor_Orcado'])

plt.figure(figsize=(8, 6))
sns.regplot(x='População', y='Somatorio_Valor_Orcado', data=df_combined, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

plt.title(f'Correlação entre População e Somatório do Valor Orçado na Paraíba (Pearson: {correlacao:.3f})')
plt.xlabel('População')
plt.ylabel('Somatório do Valor Orçado')
plt.grid(True)
plt.savefig('Correlação_pop_vlrorcado.png', dpi=300, bbox_inches='tight')
plt.show()
```

Com a análise da correlação e do coeficiente de Pearson, temos que há uma forte relação entre o crescimento demográfico da paraíba e o crescimento do orçamento público. A forte relação pode indicar que o planejamento orçamentário é sensível a mudanças demográficas, refletindo políticas públicas alinhadas com o crescimento populacional. Contudo, é válido ressaltar que correlação não implica causalidade!

### Análise dos Setores

```python
df_grouped = df.groupby(["exercicio", "funcao"])["valor_orcado"].sum().reset_index()

df_grouped["proporcao"] = (
    df_grouped.groupby("exercicio")["valor_orcado"]
    .apply(lambda x: x / x.sum())
    .reset_index(drop=True) 
)

top5_per_year = (
    df_grouped
    .sort_values(["exercicio", "valor_orcado"], ascending=[True, False])
    .groupby("exercicio")
    .head(5)
)

top5_setores = top5_per_year["funcao"].value_counts().head(5).index.tolist()
top5_per_year_filtered = top5_per_year[top5_per_year["funcao"].isin(top5_setores)]

plt.figure(figsize=(12, 6))
for setor in top5_setores:
    subset = top5_per_year_filtered[top5_per_year_filtered["funcao"] == setor]
    plt.plot(subset["exercicio"], subset["proporcao"], marker="o", linestyle="-", label=setor)

plt.xlabel("Ano")
plt.ylabel("Proporção do Orçamento")
plt.title("Evolução dos 5 Setores Mais Financiados por Ano")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout() 
plt.savefig('evolucao_setores_mais_financiados.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_agrupado = df.groupby(["exercicio", "funcao"])["valor_orcado"].sum().reset_index()

df_agrupado["proporcao"] = (
    df_agrupado.groupby("exercicio")["valor_orcado"]
    .apply(lambda x: x / x.sum())
    .reset_index(drop=True) 
)

base5_por_ano = (
    df_agrupado
    .sort_values(["exercicio", "valor_orcado"], ascending=[True, True])  # Ascending=True para pegar os menores
    .groupby("exercicio")
    .head(5)
)

base5_setores = base5_por_ano["funcao"].value_counts().head(5).index.tolist()
base5_por_ano_filtrado = base5_por_ano[base5_por_ano["funcao"].isin(base5_setores)]

plt.figure(figsize=(12, 6))
for setor in base5_setores:
    subconjunto = base5_por_ano_filtrado[base5_por_ano_filtrado["funcao"] == setor]
    plt.plot(subconjunto["exercicio"], subconjunto["proporcao"], marker="o", linestyle="-", label=setor)

plt.xlabel("Ano")
plt.ylabel("Proporção do Orçamento")
plt.title("Evolução dos 5 Setores Menos Financiados por Ano")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()  
plt.savefig('evolucao_setores_menos_financiados.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Análise de cluster dos setores/orçamentos

Preparação dos dados para clusterização:

```python
df_cluster = df.groupby(['funcao', 'exercicio'])['valor_orcado'].sum().unstack()

scaler = StandardScaler()
df_normalized = pd.DataFrame(
    scaler.fit_transform(df_cluster),
    index=df_cluster.index,
    columns=df_cluster.columns
)
```

Método do cotovelo para determinação de quantos clusters:

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_normalized)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('NUmero de Clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()
```

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(df_normalized)

df_cluster['Cluster'] = clusters
df_normalized['Cluster'] = clusters
```

```python
plt.figure(figsize=(14, 8))
for cluster in range(k):
    cluster_data = df_normalized[df_normalized['Cluster'] == cluster]
    for idx, row in cluster_data.iterrows():
        plt.plot(row.index[:-1], row.values[:-1], alpha=0.6, label=f'Cluster {cluster}' if idx == cluster_data.index[0] else "")
plt.title('Padroes de Alocação Orçamentária por Cluster')
plt.xlabel('Ano')
plt.ylabel('Valor Orçado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

cluster_summary = df_cluster.groupby('Cluster').mean().T
lines = cluster_summary.plot(ax=ax)


plt.title('Comportamento Médio dos Clusters por Ano', pad=20)
plt.ylabel('Valor Orçado (Média)')
plt.xlabel('Ano')
plt.grid(True)


cluster_descriptions = [
    "Cluster 0 :\n- Administração\n, Encargos especiais\n, Judiciário\n, Transporte",
    "Cluster 1 (Média Prioridade):\n- Educação\n- Saúde\n- Previdência Social\n, Segurança Pública",
    "Cluster 2 (Baixa Prioridade):\n- Ciência/Tecnologia\n- Agricultura\n- Cultura/Esportes"
]

handles, _ = ax.get_legend_handles_labels()

ax.legend(handles, cluster_descriptions, 
          bbox_to_anchor=(1.05, 1), 
          loc='upper left', 
          borderaxespad=0.,
          title='Descrição dos Clusters',
          frameon=False)

plt.tight_layout(rect=[0, 0, 0.85, 1])  # Reduz a área do gráfico para caber a legenda

plt.savefig('clusterss.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
for cluster in range(k):
    print(f"\nCluster {cluster} - Setores:")
    print(df_cluster[df_cluster['Cluster'] == cluster].index.tolist())
```

### Análise em anos políticos

```python
anos_eleitorais = [2014, 2016, 2018, 2020, 2022, 2024]  

df['ano_eleitoral'] = df['exercicio'].isin(anos_eleitorais)

media_geral = df.groupby('ano_eleitoral')['valor_orcado'].mean()

print("Média de investimento:")
print(f"Anos não eleitorais: R$ {media_geral[False]:,.2f}")
print(f"Anos eleitorais: R$ {media_geral[True]:,.2f}")

from scipy.stats import ttest_ind
nao_eleitoral = df[df['ano_eleitoral'] == False]['valor_orcado']
eleitoral = df[df['ano_eleitoral'] == True]['valor_orcado']
t_stat, p_value = ttest_ind(nao_eleitoral, eleitoral, equal_var=False)

print(f"\nTeste t: p-valor = {p_value:.4f}")
print("Diferença estatisticamente significativa?" , "Sim" if p_value < 0.05 else "Não")
```

Com essa análise é possível ver que, através de uma bilioteca estatística, durante os anos políticos há uma diferença de valores orçamentários. Contudo, é necessário analisar que o estado da paraíba teve uma crescente a partir de 2020, então esses dados podem interverir na análise.

```python
anos_eleitorais = [2014, 2016, 2018, 2020, 2022, 2024]

df["exercicio"] = pd.to_numeric(df["exercicio"], errors='coerce')
df["valor_orcado"] = pd.to_numeric(df["valor_orcado"], errors='coerce')

df["ano_eleitoral"] = df["exercicio"].apply(lambda x: "Eleitoral" if x in anos_eleitorais else "Não Eleitoral")

orcamento_por_ano = df.groupby(["exercicio", "ano_eleitoral"])["valor_orcado"].sum().reset_index()
media_orcamento = orcamento_por_ano.groupby("ano_eleitoral")["valor_orcado"].mean()

plt.figure(figsize=(8,5))
plt.bar(media_orcamento.index, media_orcamento.values, color=['blue', 'orange'])
plt.ylabel("Média do Orçamento")
plt.title("Comparação do Orçamento em Anos Eleitorais vs. Não Eleitorais")
plt.savefig('orçamento_medio_comparacao.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
orcamento_setor = df.groupby(["ano_eleitoral", "funcao"])["valor_orcado"].sum().reset_index()

top_setores = orcamento_setor.groupby("funcao")["valor_orcado"].sum().nlargest(5).index
orcamento_setor_filtrado = orcamento_setor[orcamento_setor["funcao"].isin(top_setores)]

plt.figure(figsize=(10,6))
for setor in top_setores:
    subset = orcamento_setor_filtrado[orcamento_setor_filtrado["funcao"] == setor]
    plt.bar(subset["ano_eleitoral"], subset["valor_orcado"], label=setor)

plt.ylabel("Orçamento Total")
plt.title("Alocação de Recursos nos Top 5 Setores - Anos Eleitorais vs. Não Eleitorais")
plt.legend()
plt.savefig('alocacao_recursos_top5setores.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Comparativo entre os elementos despesas na educação em períodos políticos e não políticos

```python
def anos_politicos(func, coluna):
    df['funcao'] = df['funcao'].str.strip()
    judiciaria = df[df['funcao'] == func]
    
    anos_politicos = judiciaria[judiciaria['exercicio'] % 2 == 0]  # Anos pares
    anos_politicos = anos_politicos[anos_politicos['exercicio'] >= 2014]  # A partir de 2014
    
    top_acoes = anos_politicos.groupby(coluna)['valor_orcado'].sum().nlargest(5).reset_index()
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x='valor_orcado', y=coluna, data=top_acoes, hue=coluna, palette='viridis', legend=False)
    plt.title(f'Top 5 {coluna} na educação em Anos Políticos (2014-2024)')
    plt.xlabel('Valor Orçado (R$)')
    plt.ylabel(coluna)
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    
   
    for i, valor in enumerate(top_acoes['valor_orcado']):
        plt.text(valor, i, f'R$ {valor:,.2f}', va='center', ha='left', fontsize=10)
    
    plt.tight_layout()
    plt.savefig(f'grafico_{func}_{coluna}_AP.png', dpi=300, bbox_inches='tight')
    plt.show()

def anos_nao_politicos(func, coluna):
    df['funcao'] = df['funcao'].str.strip()
    judiciaria = df[df['funcao'] == func]
    
    anos_nao_politicos = judiciaria[judiciaria['exercicio'] % 2 == 1]  # Anos pares
    anos_nao_politicos = anos_nao_politicos[anos_nao_politicos['exercicio'] >= 2014]  # A partir de 2014
    
    top_acoes = anos_nao_politicos.groupby(coluna)['valor_orcado'].sum().nlargest(5).reset_index()
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x='valor_orcado', y=coluna, data=top_acoes, hue=coluna, palette='viridis', legend=False)
    plt.title(f'Top 5 {coluna} em Anos não Políticos (2014-2024)')
    plt.xlabel('Valor Orçado (R$)')
    plt.ylabel(coluna)
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    
    for i, valor in enumerate(top_acoes['valor_orcado']):
        plt.text(valor, i, f'R$ {valor:,.2f}', va='center', ha='left', fontsize=10)
    
    plt.tight_layout()
    plt.savefig(f'grafico_{func}_{coluna}_ANP.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
anos_nao_politicos("EDUCACAO", "elemento_despesa")
anos_politicos("EDUCACAO", "elemento_despesa")
```

### Análise Pandemia

```python
pandemia_anos = [2020, 2021, 2022]
df_pandemia = df[df['exercicio'].isin(pandemia_anos)]

gastos_pandemia = df_pandemia.groupby('funcao')['valor_orcado'].sum().sort_values(ascending=False).head(5)

ante_pandemia = df[df['exercicio'].isin([2017, 2018, 2019])]
gastos_pre_pandemia = ante_pandemia.groupby('funcao')['valor_orcado'].sum()

variacao = ((gastos_pandemia - gastos_pre_pandemia) / gastos_pre_pandemia * 100).dropna().sort_values(ascending=False)

plt.figure(figsize=(12,6))
variacao.plot(kind='bar', color=['green' if x > 0 else 'red' for x in variacao])
plt.title('Variação Percentual dos Investimentos por Área Durante a Pandemia vs. Período Anterior')
plt.ylabel('Variação (%)')
plt.axhline(0, color='black', linestyle='--')
plt.savefig('variacao_percentual_investimentos.png', dpi=300, bbox_inches='tight')
plt.show()
```
